```python
# Define an object containing your color scheme
bg_scheme = ColorScheme()
```

```python
# Colors include red (r), blue (b), green (g), yellow (y), orange (o), and purple (p).
# Accessing those attributes returns the relevant hex code for that color:
print('The hex code for red is {}'.format(bg_scheme.r))
print('The hex code for blue is {}'.format(bg_scheme.b))
```

```python
# We can use these hex codes to color plots:
plt.figure(figsize=(6,4))
np.random.seed(196801)
data = abs(np.random.randn(5,))
plt.bar(range(len(data)),data,color=bg_scheme.b) # Here we are plotting with blue
plt.show()

plt.figure(figsize=(6,4))
np.random.seed(196801)
data = abs(np.random.randn(5,))
plt.bar(range(len(data)),data,color=bg_scheme.g) # Here we are plotting with green
plt.show()

plt.figure(figsize=(6,4))
np.random.seed(196801)
data = abs(np.random.randn(5,))
plt.bar(range(len(data)),data,color=bg_scheme.r) # Here we are plotting with red
plt.show()
```

```python
# The color scheme also contains colormaps:
# Colormaps include bwr, rwb, bwg, gwb, gwr, rwg, rpb, bpr, gbp, pbg, yrp, pry, rbp, pbr
# Note that reversing the order of the letters reverses the order of the colors
```

```python
print('BWR Scheme')
np.random.seed(19680801)
data = np.random.randn(30, 30)
plt.figure(figsize=(6,4))
plt.pcolormesh(data, cmap = bg_scheme.bwr_cm,rasterized=True, vmin=-4, vmax=4)
plt.colorbar()
plt.show()

print('GWB Scheme')
np.random.seed(19680801)
data = np.random.randn(30, 30)
plt.figure(figsize=(6,4))
plt.pcolormesh(data, cmap = bg_scheme.gwb_cm,rasterized=True, vmin=-4, vmax=4)
plt.colorbar()
plt.show()

print('GWR Scheme')
np.random.seed(19680801)
data = np.random.randn(30, 30)
plt.figure(figsize=(6,4))
plt.pcolormesh(data, cmap = bg_scheme.gwr_cm,rasterized=True, vmin=-4, vmax=4)
plt.colorbar()
plt.show()

print('RPB Scheme')
np.random.seed(19680801)
data = np.random.randn(30, 30)
plt.figure(figsize=(6,4))
plt.pcolormesh(data, cmap = bg_scheme.rpb_cm,rasterized=True, vmin=-4, vmax=4)
plt.colorbar()
plt.show()

print('GBP Scheme')
np.random.seed(19680801)
data = np.random.randn(30, 30)
plt.figure(figsize=(6,4))
plt.pcolormesh(data, cmap = bg_scheme.gbp_cm,rasterized=True, vmin=-4, vmax=4)
plt.colorbar()
plt.show()

print('PRY Scheme')
np.random.seed(19680801)
data = np.random.randn(30, 30)
plt.figure(figsize=(6,4))
plt.pcolormesh(data, cmap = bg_scheme.pry_cm,rasterized=True, vmin=-4, vmax=4)
plt.colorbar()
plt.show()

print('RPB Scheme')
np.random.seed(19680801)
data = np.random.randn(30, 30)
plt.figure(figsize=(6,4))
plt.pcolormesh(data, cmap = bg_scheme.rpb_cm,rasterized=True, vmin=-4, vmax=4)
plt.colorbar()
plt.show()
```

```python
# Bar chart with multiple colors:
n_datapoints = 5

# Select Color Scheme:
cmap = bg_scheme.gbp_cm

# Create a color list that samples from different points along the color scheme:
color_list = [cmap(x) for x in np.arange(0,1,1/n_datapoints)]

plt.figure(figsize=(6,4))
np.random.seed(196801)
data = abs(np.random.randn(5,))
plt.bar(range(len(data)),data,color=color_list) # Here we are plotting with blue
plt.show()
```
